```python
# 定义目标函数
def func(x):
    return x ** 4 - 4 * x ** 3 + 2 * x ** 2 + 3 * x + 1

# 定义目标函数的导数
def derivative(x):
    return 4 * x ** 3 - 12 * x ** 2 + 4 * x + 3
```

```python
def gradient_descent(learning_rate, num_iterations):
    x = np.random.rand()  # 随机初始化自变量x
    x_history = [x]
    for _ in range(num_iterations):
        grad = derivative(x)
        x -= learning_rate * grad
        x_history.append(x)
    return x, func(x), x_history
```

```python
learning_rate = 0.01
num_iterations = 1000
min_x, min_y, x_history = gradient_descent(learning_rate, num_iterations)
print(f"近似全局最小值点: x = {min_x}, y = {min_y}")
```

```python
# 生成用于绘制函数图像的x值
x_plot = np.linspace(-1, 4, 400)
y_plot = func(x_plot)

# 绘制函数图像
plt.plot(x_plot, y_plot, label='功能')

# 绘制梯度下降迭代过程
x_history = np.array(x_history)
y_history = func(x_history)
plt.scatter(x_history, y_history, c='r', label='梯度下降迭代')

plt.xlabel('x')
plt.ylabel('y')
plt.title('两个最小值的函数的梯度下降')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# 定义目标函数
def func(x):
    return x ** 4 - 4 * x ** 3 + 2 * x ** 2 + 3 * x + 1


# 定义目标函数的导数
def derivative(x):
    return 4 * x ** 3 - 12 * x ** 2 + 4 * x + 3


def gradient_descent(learning_rate, num_iterations, initial_x):
    x = initial_x
    x_history = [x]
    for _ in range(num_iterations):
        grad = derivative(x)
        x -= learning_rate * grad
        x_history.append(x)
    return x, func(x), x_history


learning_rate = 0.01
num_iterations = 1000
num_trials = 10  # 尝试次数
best_min_x = None
best_min_y = np.inf
all_x_history = []

for _ in range(num_trials):
    initial_x = np.random.rand()
    min_x, min_y, x_history = gradient_descent(learning_rate, num_iterations, initial_x)
    all_x_history.extend(x_history)
    if min_y < best_min_y:
        best_min_x = min_x
        best_min_y = min_y

print(f"近似全局最小值点: x = {best_min_x}, y = {best_min_y}")

# 生成用于绘制函数图像的x值
x_plot = np.linspace(-1, 4, 400)
y_plot = func(x_plot)

# 绘制函数图像
plt.plot(x_plot, y_plot, label='功能')

# 绘制梯度下降迭代过程
x_history = np.array(all_x_history)
y_history = func(x_history)
plt.scatter(x_history, y_history, c='r', label='梯度下降迭代')

plt.xlabel('x')
plt.ylabel('y')
plt.title('两个最小值的函数的梯度下降')
plt.legend()
plt.show()
```
