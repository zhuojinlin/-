```python
import torch
import torch.nn as nn


# 定义包含三层隐藏层的全连接神经网络
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # 第一层隐藏层，输入维度为10，输出维度为64（可根据实际调整）
        self.fc1 = nn.Linear(10, 64)
        # 第二层隐藏层，输入维度为64，输出维度为32（可根据实际调整）
        self.fc2 = nn.Linear(64, 32)
        # 第三层隐藏层，输入维度为32，输出维度为16（可根据实际调整）
        self.fc3 = nn.Linear(32, 16)
        # 输出层，输入维度为16，输出维度为5（对应5分类）
        self.fc4 = nn.Linear(16, 5)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = torch.relu(self.fc3(x))
        x = self.fc4(x)
        return x


# 创建网络实例
net = Net()

# 构造维度为10的单个样本输入数据，形状为 (1, 10)，这里随机生成示例数据，数据类型需为torch.float32
input_data = torch.randn(1, 10).float()

# 前向传播
output = net(input_data)
print("网络输出:", output)

# 定义真实标签，这里简单假设为一个5分类中的某一类（例如第2类，索引为1），形状为 (1,)
target = torch.tensor([1]).long()

# 定义损失函数，这里使用交叉熵损失
criterion = nn.CrossEntropyLoss()
loss = criterion(output, target)
print("单个样本损失:", loss.item())

# 反向传播
net.zero_grad()  # 清除之前的梯度信息（可选，如果之前有进行过反向传播的话）
loss.backward()

# 打印所有网络参数的偏导数（梯度）
for name, param in net.named_parameters():
    print(f"{name} 的偏导数: \n{param.grad}")
```
